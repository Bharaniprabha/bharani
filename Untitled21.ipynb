```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sb
```

```python
x=df["Review"]
mean=df["Review"].mean()
std=df["Review"].std()
z= (x-mean)/std
z
```

```python
df["z-score"]=z
z
```

```python
df.head()
```

```python
ax = sb.countplot(x="Genre", data=df,color='red')
for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.05))
plt.show()
```

```python
ax=plt.figure(figsize=(10,6))
sb.scatterplot(x='Genre',y='Review',data=df,color="green")
```
